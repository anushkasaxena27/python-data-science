```python
weather_df = pd.read_csv('weather.csv', parse_dates = ['Date.Full'],dayfirst=True)
weather_df.head()
```

Rename


```python
print(weather_df.columns.to_list())
```

```python
weather_df.columns = ['precipitation', 'date', 'month', 'week', 'year', 'city', 'code', 'location', 'state', 'avg_temp', 'max_temp', 'min_temp', 'wind_direction', 'wind_speed'] 
```

```python
weather_df.head()
```

```python
weather_df.drop(columns = ['code','location'], inplace = True) #drop columns that are not needed
```

```python
weather_df #check to see if columns were dropped
```

# Pandas sheet styling


```python
city_df = weather_df[weather_df['city'] == 'Montgomery']
city_df
```

```python
city_df.style.background_gradient(cmap='crest')
```

```python
city_df.style.bar(cmap = 'crest')
```

```python
weather_df.sort_values(by = ['state','city'], inplace = True)
```

```python
px.scatter_3d(weather_df, x = 'max_temp',y = 'wind_direction', z = 'precipitation', color = 'city')
```
